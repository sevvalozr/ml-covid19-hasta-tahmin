```python
# =====================================================
# COVID-19 HASTA TAHMİNİ PROJESİ
# Detaylı - Grafikli - Uzun Kod
# Logistic Regression + Random Forest + KNN
# =====================================================

# =========================
# KÜTÜPHANELER
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# Grafik ayarları
plt.rcParams["figure.figsize"] = (10,6)

# =========================
# VERİ SETİNİ OKU
# =========================
df = pd.read_csv("covid_19_clean_complete.csv")

print("İLK 5 SATIR")
print(df.head())

print("\nVERİ BOYUTU")
print(df.shape)

print("\nSÜTUNLAR")
print(df.columns)

# =========================
# VERİ ANALİZİ
# =========================
print("\nBİLGİLER")
print(df.info())

print("\nİSTATİSTİKSEL ÖZET")
print(df.describe())

print("\nEKSİK VERİLER")
print(df.isnull().sum())

# =========================
# EKSİK VERİ TEMİZLEME
# =========================
df = df.dropna()

# =========================
# KATEGORİK VERİLERİ SAYIYA ÇEVİR
# =========================
# Create the target variable 'covid_result' based on 'Active' cases
df['covid_result'] = (df['Active'] > 0).astype(int)

le = LabelEncoder()

# Apply LabelEncoder only to object-type columns (categorical)
for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col])

print("\nDÖNÜŞTÜRÜLMÜŞ VERİ")
print(df.head())

# =====================================================
# GRAFİKLER
# =====================================================

# 1 Hedef dağılımı
plt.figure()
sns.countplot(x="covid_result", data=df)
plt.title("COVID Sonuç Dağılımı")
plt.xlabel("Sonuç")
plt.ylabel("Kişi Sayısı")
plt.show()

# 2 Korelasyon haritası
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Korelasyon Haritası")
plt.show()

# 3 Tüm sütun histogramları
df.hist(figsize=(15,10))
plt.suptitle("Tüm Özelliklerin Histogramları")
plt.show()

# 4 Boxplot
plt.figure(figsize=(14,7))
sns.boxplot(data=df)
plt.title("Aykırı Değer Kontrolü")
plt.xticks(rotation=45)
plt.show()

# =====================================================
# GİRİŞ ÇIKIŞ AYIRMA
# =====================================================
X = df.drop("covid_result", axis=1)
y = df["covid_result"]

# =====================================================
# EĞİTİM TEST
# =====================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# =====================================================
# ÖLÇEKLENDİRME
# =====================================================
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =====================================================
# MODEL 1 LOGISTIC REGRESSION
# =====================================================
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)

acc_lr = accuracy_score(y_test, pred_lr)

print("\nLOGISTIC REGRESSION")
print("Accuracy:", acc_lr)
print(confusion_matrix(y_test, pred_lr))
print(classification_report(y_test, pred_lr))

# =====================================================
# MODEL 2 RANDOM FOREST
# =====================================================
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

acc_rf = accuracy_score(y_test, pred_rf)

print("\nRANDOM FOREST")
print("Accuracy:", acc_rf)
print(confusion_matrix(y_test, pred_rf))
print(classification_report(y_test, pred_rf))

# =====================================================
# MODEL 3 KNN
# =====================================================
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

pred_knn = knn.predict(X_test)

acc_knn = accuracy_score(y_test, pred_knn)

print("\nKNN")
print("Accuracy:", acc_knn)
print(confusion_matrix(y_test, pred_knn))
print(classification_report(y_test, pred_knn))

# =====================================================
# MODEL KARŞILAŞTIRMA GRAFİĞİ
# =====================================================
models = ["Logistic Regression", "Random Forest", "KNN"]
scores = [acc_lr, acc_rf, acc_knn]

plt.figure()
bars = plt.bar(models, scores)

plt.title("Model Accuracy Karşılaştırması")
plt.ylabel("Accuracy")
plt.ylim(0,1)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x()+0.1, yval+0.01, round(yval,3))

plt.show()

# =====================================================
# RANDOM FOREST FEATURE IMPORTANCE
# =====================================================
importance = rf.feature_importances_
features = X.columns

imp_df = pd.DataFrame({
    "Özellik": features,
    "Önem": importance
})

imp_df = imp_df.sort_values(by="Önem", ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x="Önem", y="Özellik", data=imp_df)
plt.title("Özellik Önem Sıralaması")
plt.show()

# =====================================================
# CONFUSION MATRIX GRAFİKLERİ
# =====================================================

# Logistic Regression
plt.figure()
sns.heatmap(confusion_matrix(y_test, pred_lr), annot=True, fmt="d")
plt.title("LR Confusion Matrix")
plt.show()

# Random Forest
plt.figure()
sns.heatmap(confusion_matrix(y_test, pred_rf), annot=True, fmt="d")
plt.title("RF Confusion Matrix")
plt.show()

# KNN
plt.figure()
sns.heatmap(confusion_matrix(y_test, pred_knn), annot=True, fmt="d")
plt.title("KNN Confusion Matrix")
plt.show()

# =====================================================
# SONUÇ
# =====================================================
print("\nMODEL BAŞARILARI")
print("Logistic Regression:", acc_lr)
print("Random Forest:", acc_rf)
print("KNN:", acc_knn)

best = max(scores)

if best == acc_lr:
    print("En iyi model Logistic Regression")
elif best == acc_rf:
    print("En iyi model Random Forest")
else:
    print("En iyi model KNN")

print("\nProje başarıyla tamamlandı.")
```
